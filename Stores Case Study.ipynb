```python
import pandas as pd
import numpy as np
import os as os
```

```python
# Reading the Data Set

store = pd.read_csv("C:\\Users\\Administrator\\Desktop\\Python Assignments\\DataSets\\Store.csv",encoding="Latin")

#1. Dimensions of Dataset
print("The dataset dimensions are:", store.shape)

#2. Structure of Dataset
print(store.dtypes)

#3. Column Names of Dataset 
print(store.columns)

#4. Summary of dataset
store.describe()
```

```python
#1.How many unique cities are the orders being delivered to?
print("The no. of unique cities where orders are delivered are: ", len(store.City.unique()))
print(store.City.unique())
```

```python
# 2. What is the total quantity sold in the East Region? AND 3.Find the sum of the quantity sold in the East Region

store.query("Region == 'East' ")[['Quantity']].sum()
```

```python
# 4. In the south region sort the sales in decreasing order

store.query("Region == 'South' ")[['Sales']].sort_values(by = "Sales",ascending = False)
```

```python
# 5. Find the mean of quantity for every region
store.groupby("Region")[["Quantity"]].agg(['mean'])
```

```python
#6. Find the mean of sales for every category

store.groupby("Category")[["Sales"]].agg(['mean'])
```

```python
#7. Find the max, min, sum of sales and profit for every category

store.groupby("Category")[['Sales','Profit']].agg(['max','min','sum'])
```

```python
#8. Find sum of sales and max profit for every segment

store.groupby("Segment").agg({'Sales':['sum'], 'Profit':['mean']})
```

```python
#9. For every segment find the mean of the discount

store.groupby("Segment").agg({'Discount':['mean']})
```

```python
# 10. For every segment find the most profitable customers
def get(x):
    x= x.max()
    return store[store.Profit == x]['Customer Name']

store.groupby("Segment").agg({'Profit':[get,'max']})
    
```

```python
#11. What are the top 5 sub-categories that give maximum profit

a =store.groupby("Sub-Category")[['Profit']].max()['Profit'].sort_values(ascending = False).head()
print(a)
```

```python
#12. What is the Total Sales, Quantity, Discount, Profit across Total US.

store.groupby("Country")[['Sales','Quantity','Discount','Profit']].agg(['sum'])
```

```python
#13. How many times has it taken more than 5 days from placing an order to shipping

pd.to_datetime(store['Ship Date'])
pd.to_datetime(store['Order Date'])
store['days'] = pd.to_datetime(store['Ship Date'])-pd.to_datetime(store['Order Date'])


sum(store.days > '5 days')
```

```python
# 14. Find the total number of orders in every category which has been shipped with a duration of 5 days

store[store.days < '5days'].groupby('Category')[['days']].agg(['count'])
```

```python
#15. What’s the percentage of items which has been shipped within 5 days

round(sum(store.days < '5 days')*100/len(store.days),2)
```

```python
# 16. What’s the percentage of items which has been shipped after 5 days
round(sum(store.days>'5 days')*100/len(store.days),2)
```
